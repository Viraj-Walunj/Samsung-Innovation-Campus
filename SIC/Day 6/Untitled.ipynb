```python
# Generator
# Explain generator expressions vs list comparison
def squares(n):
    for i in range(1, n+1):
        yield i * i
gen=squares(3)

print(next(gen))
print(next(gen))
print(next(gen))
# print(next(gen)) Error:StopIteration
```

```python
# Addition of two numbers using generator
def add_numbers(a, b):
    yield a + b
sum_gen = add_numbers(5, 10)
# Retrieve the single value from the generator we use next()
result = next(sum_gen)
print(f"The sum of the two numbers is: {result}")
```

```python
# Prime or not prime using generator
# This code uses a generator expression inside the all() function to check for divisors efficiently 
def is_prime(n):
    return n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))

# Usage
num = 29
print(f"{num} is prime: {is_prime(num)}")
```

```python
# Prime or not prime using generator
def is_prime(a):
    if a > 1:
        for i in range(2, int(a**0.5) + 1):
            if (a % i) == 0:
                yield "Is not prime"
                return 
        yield "Is prime"
    else:
        yield "Is not prime"

# Main part of the script to take input and run the check
try:
    number = int(input("Enter an integer: "))
    print("Your number:", number)
    for result in is_prime(number):
        print(result)
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
# Single level inheritance
# Parent Class (Base Class)
class Animal:
    def speak(self):
        print("Some generic animal sound")
# Child Class (Derived Class)
class Dog(Animal):
    def speak(self):
        print("Woof!")
# Usage
my_dog = Dog()
my_dog.speak()
```

```python
# Multilevel Inheritance
# A class is derived from another derived class
class grandfather():
    def abc(self):
        print("I am grandfather")
class father(grandfather):
    def xyz(self):
        print("i am father")
class sun(father):
    def qwe(self):
        print("i am sun")
f=sun()
f.abc ()
f.xyz()
f.qwe()
```

```python
# Polymorphism
# method overloading
class Test:
    def show(self,a):
        print(a)

    def show(self,a,b):
        print(a,b)
t=Test()
t.show(10,54)
```

```python
# addition of number using overloading
class Number:
    def __init__(self, val):
        self.val = val
    # Overloading the '+' operator
    def __add__(self, other):
        return Number(self.val + other.val)
    def __repr__(self):
        return f"Number({self.val})"
num1 = Number(10)
num2 = Number(20)
result = num1 + num2 
print(result) 
```

```python
# super()
class parent:
    def display(self):
        print("Parent Display Method")
class  Child(parent):
    def display (self):
        super().display()
        print("Child display method")
c=Child()
c.display()
```
