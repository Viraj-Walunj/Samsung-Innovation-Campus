```python
import json
data='{"title":"The old man and the sea","ISBN":"1234","Author":"Ernest Hemingway"}'
json_data = json.loads(data)

with open('book.json','w')as f:
    json.dump(json_data,f,indent='\t')
```

```python
lst =[11,22,33,44,55]
print('before pop(0)):',lst)
print('before pop(0) lst[1] =',lst[1])
lst.pop(0)
print('after pop(00 :', lst)
print('after pop(0) lt[1] =',lst[1])
print(lst[1])
```

```python
dic={0:11,1:22,2:33,3:44,4:55}

print('before pop(0):',dic.items())
print('before pop(0) dic[1]=',dic[1])
dic.pop(0)
print('after pop(0):' ,dic.items())
print('after pop(0) dic[1]=',dic[1])
```
