```python
#Exception Handling syntax
try:
    #code that might raise exception
except ExceptionType:
    # code to handle the exception
else:
    #code to run if no exception occurs
finally:
    #code to run no always
```

```python
# Handling a division by zero error
try:
    x=10/0
except ZeroDivisionError as e:
    print("zero say divide nahi ho sakta")
    b=30
    print(b)
else:
    print("Division successful")
    print(x)
finally:
    print("Execution complete")
```

```python
# Handling value error
try:
    x=int(input("enter a number"))
    b=10
    print(b)
except ValueError as e:
    print(e)
    print("Can't use string you can use only int")
else:
    print("conversion successful")
finally:
    print("Execution completed")
```

```python
# Handling multiple exception
try:
    x=int(input("enter a number:"))
    result = 10/x
    print(result)
except ValueError as e:
    print(f"ValueError: {e}")
except ZeroDivisionError as  e:
    print(f"ZeroDivisionError: {e}")
else:
    print("Division successful")
    print(result)
finally:
    print("Execution complete")
```

```python
# linear search 
def linear_search(target):
    arr =[5,8,2,9,1]
    print("Array is:",arr)
    for i in arr:
        if i == target:
            print("Found:",i)
            break
        else:
            print("Not found")
linear_search(5)
```

```python
# binary search
def binary_search(arr,target):
    low,high=0,len(arr) - 1
    while low <= high:
        mid = (low + high)

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
             low = mid + 1
        else:
            high = mid - 1
    return -1
arr=[10,20,30,40,50]
target=40
print(binary_search(arr,target))

# def bin(arr,target):
#     left,right=0,len(arr)-1
#     while left<right:
#         mid = (left + right)
```

```python
# convert a list into string using map() function
my_list = [1, 2, 3, "a", 4.5]
separator = ' ' # e.g., ', ' or '-'
result_string = separator.join(map(str, my_list))
print(result_string)
```

```python
# Reverse the string using lambda
reverse_string = lambda s:s[::-1]
original_string = "Hello , World"
reverse_string=reverse_string(original_string)
print(reverse_string)
```

```python
# shallow copy
import copy
original =[[1,2,3],[4,5,6]]
shallow = copy.copy(original)
shallow [0][0] =99
print(original)
print(shallow)
```

```python
#Deep copy
import copy
original =[[1,2,3],[4,5,6]]
deep = copy.deepcopy(original)
deep[0][0]=99
print(original)
print(deep)
```
