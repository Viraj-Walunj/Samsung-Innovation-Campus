```python
# find the factorial of 5 using function
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
# print a palindrome using a function
def is_palindrome(text):
    s = str(text)
    reversed_s = s[::-1]
    if s == reversed_s:
        print(f"'{text}' is a palindrome!")
        return True
    else:
        print(f"'{text}' is not a palindrome.")
        return False
# Example Usage
is_palindrome(121)      
is_palindrome("madam")  
is_palindrome(123)      
is_palindrome("Viraj")
```

```python
# Create a calculator usig a function
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

# Main calculator logic
def calculator():
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    while True:
        choice = input("Enter choice (1/2/3/4): ")
        if choice in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
            except ValueError:
                print("Invalid input. Please enter a valid number.")
                continue

            if choice == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")

            elif choice == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")

            elif choice == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")

            elif choice == '4':
                result = divide(num1, num2)
                print(f"{num1} / {num2} = {result}")

            next_calculation = input("Want to perform another calculation? (yes/no): ")
            if next_calculation.lower() != 'yes':
                break
        else:
            print("Invalid input. Please enter a valid choice (1/2/3/4).")

calculator()
```

```python
# prime number using function
import math
def is_prime(num):
    if number <= 1:
        return False
    for i in range(2,num):
        if number % i == 0:
            return False
    return True
n= int(input("Enter a number:"))
print("Prime" if is_prime(n) else "Not Prime")
```

```python
# Using *args 
# for handing multiple values through  SINGLE PARAMETER
def sum_all(*numbers):
    return sum(numbers)
print(sum_all(1, 2, 3, 4, 20, 30, 40)) 
```

```python
# Define and initialize the balance variable in the global scope first
# ATM machine
balance = 1000  #global variable
def withdraw(amount):
    global balance
    if amount <= balance:
        balance -= amount
        print(f"withdrawal successful. Remaining balance: {balance}")
    else:
        print("Insufficient balance.")

withdraw(200)
print(f"Balance outside function: {balance}")
```

```python
# Find the vowels from the string
def find_vowels(text):
    vowels = ['a', 'e', 'i', 'o', 'u']
    found_vowels = []
    for char in text.lower():
        if char in vowels:
            found_vowels.append(char)
    return found_vowels

my_string = "Hello World"
print(find_vowels(my_string)) 
```
