```python
# Remove Duplicate word from sentence
sentence = "python is easy and python is powerful"

# Split the sentence into a list of words
words = sentence.split()

# Convert the list to a set to remove duplicates, then back to a list
unique_words = list(set(words))

# Join the unique words back into a sentence
new_sentence = " ".join(unique_words)

print(f"Original sentence: {sentence}")
print(f"New sentence: {new_sentence}")
```

```python
# write a code for blood donation
# criteria
# if weight >=50
# age <70
# if health is good
MINIMUM_WEIGHT = 50
MAXIMUM_AGE = 70
REQUIRED_HEALTH = 'good'

weight = 45
age = 45
health_status = 'good'

if weight < MINIMUM_WEIGHT:
    print("Ineligible: Weight is below 50 kg.")
elif age >= MAXIMUM_AGE:
    print("Ineligible: Age is {age} years or older.")
elif health_status.lower() != REQUIRED_HEALTH:
    print("Ineligible: Health status is listed as '{health_status}', but must be '{REQUIRED_HEALTH}'.")
else:
    print ("Eligible: You meet all basic criteria for blood donation.")
```
