```python
import threading
import time

def print_numbers():
    for i in range(1,6):
        print("Number:",i)
        time.sleep(1)
def print_letters():
    for ch in ['A','B','C','D','E']:
        print("letter:",ch)
        time.sleep(1)

t1=threading.Thread(target=print_numbers)
t2=threading.Thread(target=print_letters)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
# Threading
import threading
import time
def worker_function():
    print("Worker thread started")
    time.sleep(2) # Simulate work
    print("Worker thread finished")
# Create a thread
thread = threading.Thread(target=worker_function)
# Start the thread
thread.start()
print("Main thread continues...")
thread.join()
print("Main thread finished")
```

```python
# handling missing values
data={
    "Name":["Aditya","Viraj","Rahul"],
    "Marks":[None,90,69]
}
df=pd.DataFrame(data)
print(df)
df["Marks"] = df["Marks"].fillna(0) #fillna() avoid errors
print("\nAfter Cleaning:")
print(df)

print("\nAverage Marks:",df["Marks"].mean())
```

```python
data={
    "Name":["Aditya","Viraj","Rahul","Utkarsh","Siddhu"],
    "Department":["IT","CS","IT","CS","IT"],
    "Marks":[None,90,69,89,67]
}
df=pd.DataFrame(data)
# Display first row 
print("1. Original Data")
print(df)
# Display first three records
print("\n2. First 3 Records")
print(df.head(3))
# Shape and data type
print("\n3. Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
#handle missing value
df["Marks"].fillna(df["Marks"].mean())
print("\n4. After Filling Missing Marks")
print(df)
# Add Result Column
def result(m):
    return "Pass" if m>= 75 else "Fail"
df["Result"] = df["Marks"].apply(result)
print("\n5. Result Column Added")
print(df)
# Filtered Passed student
passed_student = df[df["Result"] == "Pass"]
print("\n6. Filtering Passed Students")
print(passed_student)
# Sort by Marks(Descending)
sorted_df= df.sort_values("Marks",ascending=False)
print("\n7. Sorted by Marks")
print(sorted_df)
# Average Marks
avg_marks=df["Marks"].mean()
print("\n8. Average Marks:", avg_marks)
# Group by department
dept_avg=df.groupby("Department")["Marks"].mean()
print("\n9. Department Wise Average Marks")
print(dept_avg)
# Save Report to csv
sorted_df.to_csv("Student_result_report.csv", index=False)
print("\n10. Report saved as Student_result_report.csv")
```
