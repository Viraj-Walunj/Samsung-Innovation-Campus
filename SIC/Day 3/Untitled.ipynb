```python
# Repeating a list of squares of numbers then add that cubed number in a new list:
original_numbers = [1, 2, 3, 4, 5]
squares_list = []
cubes_list = []
for number in original_numbers:
    squares_list.append(number ** 2)
    cubes_list.append(number ** 3)
print("Original numbers:", original_numbers)
print("Squares list:", squares_list)
print("Cubes list:", cubes_list)
```

```python
# find the second highest number in the list (without inbuilt function)
def find_second_highest(nums):
    # Initialize with negative infinity
    high = second_high = float('-inf')

    for x in nums:
        if x > high:
            second_high = high
            high = x
        elif x > second_high and x != high:
            second_high = x
            
    return second_high

# Example usage
numbers = [12, 35, 1, 10, 34, 1]
print("Second highest:", find_second_highest(numbers))
```
