```python
# Reverseing a list without using inbuilt function

def reverse_list_new(original_list):
    reversed_list = []
    # Iterate from the last index down to the first
    for i in range(len(original_list) - 1, -1, -1):
        reversed_list.append(original_list[i])
    return reversed_list

# Example Usage:
original_list = [10, 20, 30, 40]
reversed_list = reverse_list_new(original_list)
print(reversed_list)
print(original_list)
```

```python
# Reverseing a string without using inbuilt function

def reverse_for_loop(s):
    reversed_s = ""
    for char in s:
        reversed_s = char + reversed_s
    return reversed_s

# Example
input_str = "Python"
print(f"Original string: {input_str}")
print(f"Reversed string: {reverse_for_loop(input_str)}")
```

```python
# Accesing value in a tuple using indexing / Fetch the specific element using indexing
my_tuple = ("apple", "banana", "cherry", "orange")

# Accessing the first element (index 0)
first_item = my_tuple[0]
print(first_item)

# Accessing the third element (index 2)
third_item = my_tuple[2]
print(third_item)
```

```python
# Check existence of an item in a python tuple using if else
my_tuple = ('apple', 'banana', 'cherry', 'date')
item_to_find = 'banana'

if item_to_find in my_tuple:
    print(f"{item_to_find} exists in the tuple.")
else:
    print(f"{item_to_find} does not exist in the tuple.")
```
