```python
# Lambda Function
# Square of a number
# lambda argument(Ex: x) : expression(Ex: x * x)
square = lambda x: x * x #from : the code starts after the : it is the expression(logic)
print(square(5))
```

```python
# lambda vs normal function
# normal function
def add(a,b):
    return a + b


# lambda function(same as above but in 1 line)
add = lambda a,b :a + b
```

```python
# pass multiple value return in lambda function
add = lambda x, y: (x + y, x - y)
a, s = add(x=4, y=5) # Use the equals sign for keyword arguments
print(a)
print(s)
```

```python
# default keyword arg
a=lambda x,y=2:x+y
print(a(2))
```

```python
# map() function
numbers =[1,2,3,4]
# map(function,iterable)
result = list(map(lambda x: x * x, numbers))
print(result)
```

```python
# convert string to uppercase
name = ["ram","shyam","geeta"]
uppercase= list(map(lambda x: x.upper(),name))
print(uppercase)
```

```python
# filter function
marks = [30, 40, 60, 70, 67, 66]
def failing(scored): 
    return scored < 60
result = filter(failing, marks)
print("failing score:", list(result))
```

```python
# Even numbers
numbers=[1,2,3,4,5,6]
result=list(filter(lambda x: x % 2==0,numbers))
print(result)
```

```python
# find odd/ even numbers using lambda function
check_number = lambda num: "even" if num % 2 == 0 else "odd"

number =12
print(f"{number} is an {check_number(number)} number.")

number =15
print(f"{number} is an {check_number(number)} number.")
```

```python
# DECORATOR (denote by @ symbol before the function ex:@my_decorartor)
# to call any function between any another or between to function 
def my_decorator(hello):
    def wrapper():
        print("Before decorartor")
        hello()
        print("After decorator")
    return wrapper

@my_decorator
def say_hello():
    print("Hello World")
say_hello()
```

```python
def my_deco(say_hello):
    def wrapper():
        print("Before decorartor")
        say_hello()
        print("After decorator")
    return wrapper

def second_deco(say_hello):
    def wrapper():
        print("== Extra functionality added ==")
        say_hello()
        print("== End of Extra functionality ==")
    return wrapper

@second_deco
@my_deco
def say_hello():
    print("Hello Team")
say_hello()
```

```python
# converting lowercase to uppercase using decorator
def upper_case_decorator(func):
    def wrapper():
        result = func()
        modified = result.upper()
        return modified
    return wrapper
@upper_case_decorator
def greet():
    return "hello team, welcome to django class"
greet()
```

```python
# addition of two numbers using decorator
def log_addition(func):
    def wrapper(a, b):
        print(f"Adding {a} and {b} is")
        return func(a, b)
    return wrapper

@log_addition
def add_numbers(a, b):
    return a + b
print(f"Result: {add_numbers(5, 10)}")
```

```python
def login_required(func):
    def wrapper(user):
        if not user.get("is_logged_in"):
            print("Access Denied! Please log in.")
            return None
        return func(user)
    return wrapper
@login_required
def dashboard(user):
    print("Welcome "  + user["name"] +" to your dashboard!")

user1 = {"name": "Nikhil", "is_logged_in": True}
user2  ={"name" : "Guest", "is_logged_in": False}
dashboard(user1)
dashboard(user2)
```
